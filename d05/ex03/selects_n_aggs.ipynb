## Imports

```python
import pandas as pd
```

## Read data
and set CarNumbers as index

```python
df = pd.read_json('../data/auto.json', orient='records')
df
```

```python
df.set_index('CarNumber', inplace=True)
df
```

## Making Selects

Display the rows only where the fines are more than 2,100

```python
df.loc[df.Fines > 2100]
```

Display the rows only where the fines are more than 2,100 and the refund
equals 2

```python
df.loc[(df.Fines > 2100) & (df.Refund == 2)]
```

Display the rows only where the models are from the list: [’Focus’, ’Corolla’]

```python
lst = ['Focus', 'Corolla']
df.loc[df.Model.apply(lambda x: x in lst)]
```

Display the rows only where the car number is from the list: [’Y7689C197RUS’,
’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]

```python
lst = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df.loc[map(lambda x: x in lst, df.index)]
```

## Making aggregations with the make and the model

Display the median fines grouped by the make

```python
df.groupby('Make').Fines.agg(['median'])
```

Display the median fines grouped by the make and the model

```python
df.groupby(['Make', 'Model']).Fines.agg(['median'])
```

Display the number of fines grouped by the make and the model in order to
understand if we can trust the median values

```python
df.groupby(['Make', 'Model']).Fines.agg(['count'])
```

Display the minimum and the maximum fines grouped by the make and the
model in order to better understand the variance

```python
df.groupby(['Make', 'Model']).Fines.agg(['min', 'max'])
```

Display the standard deviation of the fines grouped by the make and the model
in order to better understand the variance

```python
df.groupby(['Make', 'Model']).Fines.agg(['std'])
```

## Making aggregations with the car number

Display the car numbers grouped by the number of the fines in descending
order, we want to find those who most often violated the law

```python
df.groupby(['CarNumber']).Fines.agg(['count']).rename(columns={'count':'n_fines'}).sort_values('n_fines', ascending=False).groupby(['n_fines', 'CarNumber'], sort=False).first().iloc[::]
```

Select from the initial dataframe all the rows corresponding to the top-1 car
number, we want to zoom in a little bit

```python
df.loc[df.index[0]]
```

Display the car numbers grouped by the sum of the fines in descending order,
we want to find those who paid the most

```python
df.groupby(['CarNumber']).Fines.agg(['sum']).rename(columns={'sum':'sum_fines'}).sort_values('sum_fines', ascending=False).groupby(['sum_fines', 'CarNumber'], sort=False).first().iloc[::]
```

Select from the initial dataframe all the rows corresponding to the top-1 car
number, we want to zoom in a little bit

```python
df.loc[df.index[0]]
```

Display a table that answers the question: are there any car numbers that were
connected to different models?

```python
df.groupby(['CarNumber']).Model.agg(['nunique']).rename(columns={'nunique':'uniq_models'}).sort_values('uniq_models', ascending=False).groupby(['uniq_models', 'CarNumber'], sort=False).first().iloc[::]
```
